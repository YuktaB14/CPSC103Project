#### Design data definitions

```python
from cs103 import *
from typing import NamedTuple, List
import matplotlib.pyplot as plt
import csv

from typing import Optional   

Dependents = Optional[int]  # in range[0, 2] 
# interp. 3 or more than 3 dependents (None), or dependents are 0, 1, and 2.

D0 = None
D1 = 0
D2 = 1
D3 = 2

@typecheck
# template based on One-Of (2 cases), Atomic Distinct, and Atomic Non-Distinct
def fn_for_dependents(d: Dependents) -> ...:
    if d is None:
        return ...
    else:
        return ...(d)
    

ApplicantInfo = NamedTuple('ApplicantInfo', [('married', bool), #True if applicant is married, else False
                                             ('dependents', Dependents),
                                             ('applicantincome', float), #in thousands
                                             ('coapplicantincome', float), #in thousands
                                             ('loanamt', float), #in thousands
                                             ('loanterm', int), #in months
                                             ('credithist', bool), #True if applicant has credit history, else False
                                             ('status', bool)]) #True if home loan is approved, else False


#interp. each row represents information pertaining to the applicant's home loan application.

A1 = ApplicantInfo(True, D2, 5649, 0, 44, 360, True, True)
A2 = ApplicantInfo(False, D1, 3167, 0, 74, 360, True, False)
A3 = ApplicantInfo(False, D1, 3167, 0, 74, 360, False, False)



@typecheck
def fn_for_applicantinfo(ai: ApplicantInfo) -> ...: #template based on compound and reference
    return ...(ai.married,
               fn_for_dependents(ai.dependents),
               ai.applicantincome,
               ai.coapplicantincome,
               ai.loanamt,
               ai.loanterm,
               ai.credithist,
               ai.status)
    
    
    
# List[ApplicantInfo]
# interp. a list of ApplicantInfo

LOAI0 = []
LOAI1 = [A1]
LOAI2 = [A2]
LOAI3 = [A1, A2]
LOAI4 = [A1, A2, A3]

@typecheck
def fn_for_loai(loai: List[ApplicantInfo]) -> ...:
    # description of the accumulator
    acc = ...      # type: ...
    for ai in loai:
        acc = ...(fn_for_applicantinfo(ai), acc)
 
    return ...(acc)
```

Complete these steps in the code cell below. You will likely want to rename the analyze function so that the function name describes what your analysis function does.

```python
# Functions

@typecheck
def handleBool(val: str, trueVal: str) -> bool:
    """
    checks if the row value (val) is equal to the trueValue and returns True if that is the case else False
    """
    return val == trueVal

start_testing()
expect(handleBool("Yes", "Yes"), True)
expect(handleBool("No", "Yes"), False)
summary()

@typecheck
def handleDependents(dependents: str) -> Dependents:
    """
    change dependents to Dependents data type
    """
    if dependents == "3+":
        return None
    if dependents == "":
        return 0
    return int(dependents)

start_testing()
expect(handleDependents(""), 0)
expect(handleDependents("1"), 1)
expect(handleDependents("2"), 2)
expect(handleDependents("3+"), None)
summary()

@typecheck
def handleCreditHist(creditHist: str) -> bool:
    """
    Changes credit hist string from data to a boolean value
    """
    if creditHist == "1":
        return True
    return False

start_testing()
expect(handleCreditHist("1"), True)
expect(handleCreditHist("0"), False)
expect(handleCreditHist(""), False)
summary()
    
@typecheck
def read(filename: str) -> List[ApplicantInfo]:
    """    
    reads information from the specified file and returns List[ApplicantInfo]
    """
    #return []  #stub
    # Template from HtDAP
    # loai contains the result so far
    loai = [] # type: List[ApplicantInfo]

    with open(filename) as csvfile:
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            if not(row[2] == '' or row[8] == '' or row[9] == ''):
                row[2] = handleBool(row[2], "Yes")
                row[3] = handleDependents(row[3])
                row[10] = handleCreditHist(row[10])
                row[12] = handleBool(row[12], "Y")
                ai = ApplicantInfo(row[2], row[3], float(row[6]), float(row[7]), float(row[8]), int(row[9]), row[10], row[12])
                loai.append(ai)
    
    return loai

start_testing()
expect(read("sample_empty.csv"), [])
expect(read("sample_1missing.csv"), [ApplicantInfo(married=True, dependents=1, applicantincome=4583, coapplicantincome=1508, loanamt=128, loanterm=360, credithist=True, status=False)])
expect(read("sample_3missing.csv"), [ApplicantInfo(married=False, dependents=1, applicantincome=3596.0, coapplicantincome=0.0, loanamt=100.0, loanterm=240, credithist=False, status=True)])
summary()

# Functions
# Planning: 
# 1. Filter Traditional & Non traditional family structure: Traditional is married + Dependent>=1,
#    Non traditional is everything else
# 2. Filter Rejection & Approval, traditional & non traditional family structure
# 3. Filter Financial status (Debt Income Ration)

# helper function of traditional family, dependents 
@typecheck
def is_traditional_dependents(d: Dependents) -> bool:
    """
    return the number of dependents. Traditional dependents should be at least one dependents. 
    """
    #return False # stub
    #template from ApplicantInfo
    if d is None:
        return False
    else:
        if d >= 1:
            return True
        else:
            return False

start_testing()     
expect(is_traditional_dependents(A1.dependents),True)
expect(is_traditional_dependents(A2.dependents),False)
summary()

# helper function of traditional family, married 
@typecheck
def is_traditional(ai:ApplicantInfo) -> bool:
    """
    return True if the applicant is get married.Otherwise return False.
    """
    #return True # stub
    # Template from ApplicantInfo
    return (ai.married and is_traditional_dependents(ai.dependents))
            

start_testing()
expect(is_traditional(A1),True)
expect(is_traditional(A2),False)
summary()

# Filter Traditional family structure
@typecheck
def traditional_family(loai:List[ApplicantInfo]) -> List[ApplicantInfo]:
    """
    return the applicants who is a traditional family structure, which is married and at least one dependents. 
    """
    #return [] #stub
    #template based on List[ApplicantInfo]
    traditional_family = []
    for ai in loai:
        if is_traditional(ai):
            traditional_family.append(ai)
    return traditional_family
    

start_testing()
expect(traditional_family(LOAI1),[A1])
expect(traditional_family(LOAI3),[A1])
summary()


# Filter Non Traditional family structure 
@typecheck
def non_traditional_family(loai: List[ApplicantInfo]) -> List[ApplicantInfo]:
    """
    return the applicants who is a non traditional structure, which is not married and no any dependents. 
    """
    #return [] # stub
    #template based on List[ApplicantInfo]
    non_traditional_family = []
    for ai in loai:
        if not(is_traditional(ai)):
             non_traditional_family.append(ai)
    return  non_traditional_family

start_testing()
expect(non_traditional_family(LOAI2),[A2])
expect(non_traditional_family(LOAI3),[A2])
summary()

# I need to write a helper function to prove whether it is false or True 
# Filter rejection
@typecheck
def number_rejection_applicants(loai: List[ApplicantInfo]) -> int: 
    """
    return the number of applicants who are rejected to get a home loan. 
    """
    #return 0 # stub
    count = 0
    for ai in loai:
        if not ai.status:
            count += 1
    return count 
    
    
# example of rejection 
start_testing()
expect(number_rejection_applicants(LOAI3),1)
expect(number_rejection_applicants(LOAI1),0)
summary()     
       

# Filter approval 
@typecheck
def number_approval_applicants(loai: List[ApplicantInfo])->int: 
    """
    return the number of applicants who are approved to get a home loan. 
    """
    #return 0 # stub
    count = 0
    for ai in loai:
        if ai.status:
            count += 1 
    return count 

    
start_testing()
expect(number_approval_applicants(LOAI2),0)
expect(number_approval_applicants(LOAI3),1)
summary()

def filter_by_credit_hist(loai: List[ApplicantInfo], has_credit: bool) -> List[ApplicantInfo]:
    """
    Filters loai based on whether the applicant has a credit history
    """
    #return [] # stub
    acc = [] #type: List[ApplicantInfo]
    for ai in loai:
        if ai.credithist and has_credit:
            acc.append(ai)
        if (not ai.credithist) and (not has_credit):
            acc.append(ai)
    return acc

start_testing()
expect(filter_by_credit_hist([], True), [])
expect(filter_by_credit_hist(LOAI3, False), [])
expect(filter_by_credit_hist(LOAI4, True), [A1, A2])
expect(filter_by_credit_hist(LOAI4, False), [A3])
summary()

@typecheck
def create_bar_chart_approval(loai: List[ApplicantInfo]) -> None:
    """
    Creates the number of approvals bar chart
    """
    x = ["Good Credit History", "Bad Credit History"]
    have_credit_hist_list = filter_by_credit_hist(loai, True)
    not_have_credit_hist_list = filter_by_credit_hist(loai, False)
    
    list_traditional_good = traditional_family(have_credit_hist_list)
    list_non_traditional_good = non_traditional_family(have_credit_hist_list)
    list_traditional_bad = traditional_family(not_have_credit_hist_list)
    list_non_traditional_bad = non_traditional_family(not_have_credit_hist_list)
    
    num_approval_traditional_good = number_approval_applicants(list_traditional_good)
    num_approval_traditional_bad = number_approval_applicants(list_traditional_bad)
    num_approval_non_traditional_good = number_approval_applicants(list_non_traditional_good)
    num_approval_non_traditional_bad = number_approval_applicants(list_non_traditional_bad)
    
    y1 = [num_approval_traditional_good, num_approval_traditional_bad]
    y2 = [num_approval_non_traditional_good, num_approval_non_traditional_bad]
    # Template based on visualization
    # create the first bar chart
    rects1 = plt.bar(x, 
                     y1,
                     color='b')                     # set the colour (here, blue)
    rects2 = plt.bar(x,
                     y2,
                     bottom=y1,
                     color='r')

    # set the labels for the x-axis, y-axis, and plot title
    plt.xlabel('Credit History')
    plt.ylabel('Number of Approvals')
    plt.title('Visualisation of the number of approvals')
    plt.legend(["Traditional Family", "Non-Traditional Family"])
    
    # show the plot
    plt.show()
    
    # by default, Python returns None if it gets to the end of a function and there is no call to return
    # so we could have omitted the next line of code. It also returns None when there is a return 
    # statement that does not explicitly return a value (like we have here)
    return

start_testing()
expect(create_bar_chart_approval([]), None)
expect(create_bar_chart_approval(LOAI4), None)
summary()

@typecheck
def create_bar_chart_rejection(loai: List[ApplicantInfo]) -> None:
    """
    Creates the number of rejections bar chart
    """
    x = ["Good Credit History", "Bad Credit History"]
    have_credit_hist_list = filter_by_credit_hist(loai, True)
    not_have_credit_hist_list = filter_by_credit_hist(loai, False)
    
    list_traditional_good = traditional_family(have_credit_hist_list)
    list_non_traditional_good = non_traditional_family(have_credit_hist_list)
    list_traditional_bad = traditional_family(not_have_credit_hist_list)
    list_non_traditional_bad = non_traditional_family(not_have_credit_hist_list)
    
    num_approval_traditional_good = number_rejection_applicants(list_traditional_good)
    num_approval_traditional_bad = number_rejection_applicants(list_traditional_bad)
    num_approval_non_traditional_good = number_rejection_applicants(list_non_traditional_good)
    num_approval_non_traditional_bad = number_rejection_applicants(list_non_traditional_bad)
    
    y1 = [num_approval_traditional_good, num_approval_traditional_bad]
    y2 = [num_approval_non_traditional_good, num_approval_non_traditional_bad]
    # Template based on visualization
    # create the first bar chart
    rects1 = plt.bar(x, 
                     y1,
                     color='b')                     # set the colour (here, blue)
    rects2 = plt.bar(x,
                     y2,
                     bottom=y1,
                     color='r')

    # set the labels for the x-axis, y-axis, and plot title
    plt.xlabel('Credit History')
    plt.ylabel('Number of Rejections')
    plt.title('Visualisation of the number of rejections')
    plt.legend(["Traditional Family", "Non-Traditional Family"])
    
    # show the plot
    plt.show()
    
    # by default, Python returns None if it gets to the end of a function and there is no call to return
    # so we could have omitted the next line of code. It also returns None when there is a return 
    # statement that does not explicitly return a value (like we have here)
    return

start_testing()
expect(create_bar_chart_rejection([]), None)
expect(create_bar_chart_rejection(LOAI4), None)
summary()

@typecheck
def create_charts(loai: List[ApplicantInfo]) -> None:
    """
    Creates 2 bar charts
    1. Number of approvals
    2. Number of rejctions
    """
    create_bar_chart_approval(loai)
    create_bar_chart_rejection(loai)
    return

start_testing()
expect(create_charts([]), None)
expect(create_charts(LOAI4), None)
summary()

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, creates the bar charts,
    returns the result
    """
    # return ... # stub (make sure you update the ... with a value of the correct type)
    
    # Template from HtDAP, based on composition
    return create_charts(read(filename))

start_testing()
expect(main("sample_1missing.csv"), None)
expect(main("sample_empty.csv"), None)
summary()
```
